```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import plotly.offline as py
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode
init_notebook_mode(connected=True) 
%matplotlib inline
```

```python
# load data
data_scrubbed = pd.read_csv('scrubbed.csv', low_memory = False)
display(data_scrubbed.head())
```

## Peek the data
Here we check the format and attributes about the data using `shape`, `head`, `info`, ..., etc. 

```python
data_scrubbed.shape
```

```python
# check data type of each attribute
data_scrubbed.info()
```

```python
# check missing data
print(data_scrubbed.isnull().sum())
```

Now we find out that the attributes `state`, `country`, and `shape` contain most missing data. Instead of getting rid of them, we replace those null values with the string `'unknown'`.

```python
# fill null values by 'unknown'
data_scrubbed.fillna('unknown', inplace = True)

display(data_scrubbed.isnull().sum())
```

Besides showing exact time (`datetime`), we add the columns `sight_year` and `season` from `datetime` and `date_post_year` from `date posted`.

```python
def extract_year(x):
    x = x.split(" ")[0]
    x = x.split("/")[2]
    x = int(x)
    return x

def conv_season(x):
    x = int(x.split("/")[0])
    
    if x in range(3,6):
        return "Spring"
    if x in range(6,9):
        return "Summer"
    if x in range(9,12):
        return "Autumn"
    if x == 12 or x == 1 or x == 2:
        return "Winter"

data_scrubbed["sight_year"] = data_scrubbed['datetime'].apply(extract_year)
data_scrubbed["date_posted_year"] = data_scrubbed['date posted'].apply(extract_year)
data_scrubbed["season"] = data_scrubbed['datetime'].apply(conv_season)
```

```python
# sight site
sights = data_scrubbed.groupby('country')['shape'].size().sort_values(ascending=False)
display(sights)
sights.plot(kind='barh', figsize = (20, 8))
plt.xlabel('Count')
plt.title('Sight Site')
plt.show()
```

According to the horizontal bar chart, the majority of the sighting places occur in USA. Thus, for the latter part we only focus on USA.

- **Sight years and date-posted years**

```python
data_us = data_scrubbed[data_scrubbed.country == "us"]

plt.subplots(figsize=(22,10))
plt.subplot(2,1,1)
plt.title("Sight Count by Years")
data_us['sight_year'].value_counts().sort_index().plot(kind = 'bar')
plt.xlabel("Year")
plt.subplots(figsize=(22,10))

plt.subplot(2,1,2)
plt.title("Posting the Cases' Counts by Years")
data_us['date_posted_year'].value_counts().sort_index().plot(kind = 'bar')
plt.xlabel("Post Year")
plt.xticks(rotation=0)
plt.show()
```

1. From the histograms of sight counts and posting counts, we could find out the increasing trend of finding UFOs. Especially after 2000, it grows sharply.
2. In 2012 and 2013, the sighting count and posting count reach the peak, almost 7,000 cases happened at that time.

- **Which seasons are sighted?**

```python
# find out the seasons that UFOs have been sighted 
plt.subplots(figsize = (20, 12))
plt.subplot(2, 1, 1)
data_us['season'].value_counts().plot(kind = 'bar')
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.title("Sight Season (Total)")

plt.subplot(2, 1, 2)
(data_us['season'].value_counts() / (data_us['season'].value_counts().sum())).plot(kind = 'bar')
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.title("Sight Season (Average)")
plt.show()
```

Apart from years, we observe that summer is the season that people witness UFOs the most followed by autumn.

- **Sight counts in each state of USA**

```python
# sight freq in each state
states_us = data_us["state"].value_counts().index
states_freq = data_us["state"].value_counts().values
states_us = [i.upper() for i in states_us]

plt.subplots(figsize = (24,8))
sns.barplot(states_us, states_freq)
plt.xticks(rotation = 45, fontsize = 16)
plt.yticks(fontsize = 20)
plt.xlabel('State')
plt.ylabel('Count')
plt.title('Sight Count')
plt.show()
```

California (CA) is the state where people spot UFOs the most. The number is 2 times more than the one in Washington (WA), which is in the second place.

- **USA mapplot of sight count**

```python
data = [dict(
        type='choropleth',
        locations = states_us,
        z = states_freq,
        locationmode = 'USA-states',
        text = "times",
        marker = dict(line = dict (color = 'rgb(255,255,255)', width = 2)),
        colorbar = dict(title = "Sight rates by states"))]

layout = dict(
        title = 'UFO sight rates from USA',
        geo = dict(scope='usa', projection=dict(type='albers usa'), showlakes = True))


fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```

- **How many seconds UFOs have been seen?**

```python
# transform time descriptions into numbers in seconds
duration_sec = [i for i in data_us["duration (seconds)"].value_counts()]
duration_sec_list = []
for i in duration_sec:
    if i in range(0,16):
        duration_sec_list.append("0-15")
    if i in range(15,31):
        duration_sec_list.append("15-30")
    if i in range(31,61):
        duration_sec_list.append("30-60")
    if i in range(60,121):
        duration_sec_list.append("60-120")
    if i in range(120,241):
        duration_sec_list.append("120-240")
    if i > 240:
        duration_sec_list.append(">240")
duration_sec_list = pd.Series(duration_sec_list)
di = duration_sec_list.value_counts().index
dv = duration_sec_list.value_counts().values

plt.subplots(figsize = (22,8))
sns.barplot(di, dv)
plt.xlabel("Time (sec)")
plt.xticks(fontsize = 20)
plt.ylabel("Count")
plt.yticks(fontsize = 20)
plt.title('Duration being Found')
plt.show()
```

1. From the chart above, almost all durations that UFOs being found are in the range from 0 to 15 seconds. 
2. We are a little bit suprised that roughly 50 times that people spot UFOs in more than 240 seconds a time.

- **What do the shapes look like?**

```python
# find out what kinds of UFO shapes have been sighted 
data_us['shape'].value_counts().plot(kind = 'bar', figsize = (18, 8))
plt.xticks(rotation = 90, fontsize = 15)
plt.ylabel('Count')
plt.title("UFOs' Shape being Found")
plt.show()
```

1. Most UFOs being witnessed are in the light shape followed by the triangle shape.
2. The number of light shape is 2 times more than the triangle shape.

- **Most used words in comments**

```python
words = [i for i in data_us.comments.dropna()]
words = " ".join(words)

plt.subplots(figsize=(28,12))
wordcloud = WordCloud(background_color='black', width=2048, height=1024).generate(words)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

1. From the word cloud, we speculate that people mostly spot the UFOs of the light shape, corresponding to the previous histogram that light-shaped UFOs are being found the most.
2. People describe the UFOs' colors are possibly bright, white, or red.
